# Merging FEC Data

Source(s): 
- [FEC Bulk Data (2000-2020)](https://www.fec.gov/data/browse-data/?tab=bulk-data)

Going through each directory and merging the induvidual files corresponding to each 2 year period.

```
folder = 'Operation Expenses'
_path = os.path.join(FEC_path, folder)
files = os.listdir(_path)
header = pd.read_csv(os.path.join(_path, 'header.csv'))
df = pd.concat(
      [pd.read_csv(os.path.join(_path, f), delimiter = '|', header = None)
        for f in files if 'csv' not in f])
```

```
df.drop(columns = [25], inplace = True)
df.columns = header.columns
```

```
_new_path =  os.path.join(data_path, 'FEC_Clean', f"{folder}.csv")
df.to_csv(_new_path, encoding='utf-8', index=False)
```

```
for folder in os.listdir(FEC_path):
  _path = os.path.join(FEC_path, folder)
  files = os.listdir(_path)
  header = pd.read_csv(os.path.join(_path, 'header.csv'))
  if folder == 'Operation Expenses':
    df = pd.concat(
      [pd.read_csv(os.path.join(_path, f), delimiter = '|', header = None)
        for f in files if 'csv' not in f])
    df.columns = header.columns
    _new_path =  os.path.join(data_path, 'FEC_Clean', f"{folder}.csv")
    df.to_csv(_new_path, encoding='utf-8', index=False)
```

# Polling Data

Source(s):
- [The Economist: Potus Model](https://github.com/TheEconomist/us-potus-model/tree/master/data)
- [FiveThirtyEight](https://github.com/fivethirtyeight/data/tree/master/pollster-ratings)

```
poll_dem_path = os.path.join(data_path, 'Polling and Demographics')
_path = os.path.join(poll_dem_path, 'FiveThirtyEight')
```

```
polls = pd.read_csv(os.path.join(_path, 'POLLS.csv'))
polls['electiondate'] = pd.to_datetime(polls['electiondate'])
polls['polldate'] = pd.to_datetime(polls['polldate'])
polls = polls.query("type_simple == 'Pres-P' & polldate < electiondate")

good_cols = ['pollster_rating_id', 'location', 'polldate','samplesize', 'cand1_name', 'cand1_pct', 'cand2_name', 'cand2_pct', 'electiondate']
polls = polls[good_cols]
```
